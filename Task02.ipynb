```python
import pandas as pd
import numpy as np

INPUT_PATH  = "WA_Fn-UseC_-Telco-Customer-Churn.csv"
OUTPUT_PATH = "telco_churn_cleaned.csv"

df = pd.read_csv(INPUT_PATH)
print(f"Loaded  : {df.shape[0]:,} rows × {df.shape[1]} columns")

df["TotalCharges"] = df["TotalCharges"].str.strip()
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

blanks = df["TotalCharges"].isna().sum()
print(f"TotalCharges blanks found: {blanks}")

df.loc[df["TotalCharges"].isna(), "TotalCharges"] = 0.0

df["SeniorCitizen"] = df["SeniorCitizen"].map({0: "No", 1: "Yes"})

df["ChurnFlag"] = df["Churn"].map({"Yes": 1, "No": 0})

yes_no_cols = [
    "Partner", "Dependents", "PhoneService", "PaperlessBilling",
    "Churn", "SeniorCitizen"
]
for col in yes_no_cols:
    df[col] = df[col].str.strip().str.title()

df["AnomalyFlag"] = ((df["tenure"] == 0) & (df["TotalCharges"] > 0)).astype(int)
anomalies = df["AnomalyFlag"].sum()

print(f"Final shape : {df.shape[0]:,} rows × {df.shape[1]} columns")

df.to_csv(OUTPUT_PATH, index=False)
print(f"\nCleaned file saved → {OUTPUT_PATH}")
```
